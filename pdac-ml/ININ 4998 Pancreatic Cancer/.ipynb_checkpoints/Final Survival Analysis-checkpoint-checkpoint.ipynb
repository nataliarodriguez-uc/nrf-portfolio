# Feature Selection (Actual)

## Feature Selection Funtions

```python
def select_k_best_x_data(k):
    fs= Pipeline([
        ("scaler", StandardScaler()),
        ("feature_select", SelectKBest(f_classif, k=k)),
    ])
    fit=fs.fit(X, Y.values.ravel())
    Xnew = fit.transform(X)
    Xnew_names = [col[0]  for col in zip(X.columns,fs.named_steps['feature_select'].get_support()) if col[1]]
    Xreduced = pd.DataFrame(Xnew,columns=Xnew_names)
    return Xreduced
```

```python
#Feature Extraction with RFE
from sklearn.feature_selection import RFE, RFECV
from sklearn.linear_model import LogisticRegression
from sklearn.decomposition import PCA

def rfe_x_data(n):
    #feature extraction
    model = LogisticRegression()
    fs= Pipeline([
        ("scaler", StandardScaler()),
        ("feature_select", PCA()),
    ])
    fit = fs.fit(X, Y.values.ravel())
    Xnew = fit.transform(X)
#     Xnew_names = [col[0]  for col in zip(X.columns,fs.named_steps['feature_select'].get_support()) if col[1]]
#     Xreduced = pd.DataFrame(Xnew,columns=Xnew_names)
    return Xnew
```

```python
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.feature_selection import SelectFromModel

def extra_trees_classifier(n):
    fs= Pipeline([
        ("scaler", StandardScaler()),
        ("feature_select", SelectFromModel(ExtraTreesClassifier(n_estimators=n)) ),
    ])
    fit = fs.fit(X, Y)
    Xnew = fit.transform(X)
    Xnew_names = [col[0]  for col in zip(X.columns,fs.named_steps['feature_select'].get_support()) if col[1]]
    Xreduced = pd.DataFrame(Xnew,columns=Xnew_names)
    return Xreduced
```

```python
from sklearn.feature_selection import SelectKBest, f_classif 
from sklearn.feature_selection import chi2 
from numpy import set_printoptions
import time

t0 = time.time()
def feature_selected_x_data():
    
    x_data_reduced = []
    
    for k in [75, 100, 130]:
        x_reduced = select_k_best_x_data(k)
        x_data_reduced.append( (x_reduced, "select_k_best_" + str(k) ) )
   
    x_reduced = rfe_x_data(1)
    x_data_reduced.append( (x_reduced, "pca" ) )

    for n in [75,100,130]:
        x_reduced = extra_trees_classifier(n)
        x_data_reduced.append( (x_reduced, "extra_trees_classifier" + str(n) ) )
    
        
    return x_data_reduced
feature_selected_x_data()
t1 = time.time() - t0
print("Time elapsed (seconds): ", t1)
```

```python
def save_plot_roc_curve(fpr, tpr, figname, label=None):
    plt.plot(fpr, tpr, linewidth=2, label=label)
    plt.plot([0,1],[0,1],'k--') #dashed diagonal
    plt.xlabel("False positive Rate")
    plt.ylabel("True Positive Rate (Recall)")
    plt.savefig("./figs/plot_%s.png" % ''.join(figname.split()).lower())
    plt.clf()
```

```python
def createLogisticRegression():    
    logreg= Pipeline([
        ("scaler", StandardScaler()),
        ("logistic_reg", LogisticRegression(max_iter=500)),
    ])
    return logreg
```

```python
from sklearn.ensemble import RandomForestClassifier
def createSVC():    
    pipe= Pipeline([
        ("scaler", StandardScaler()),
        ("classifier", SVC(decision_function_shape="ovo")),
    ])
    return pipe
```

```python
from sklearn.ensemble import RandomForestClassifier
def createRandomForestClassifier():    
    pipe= Pipeline([
        ("scaler", StandardScaler()),
        ("classifier", RandomForestClassifier(n_estimators=100, max_depth=3, random_state=0)),
    ])
    return pipe
```

```python
def predict_with_data(data_x, data_y, classifier, figname=''):
    #x_train,x_test,y_train,y_test=train_test_split(data_x,data_y,test_size=0.25,random_state=0)
    classifier.fit(data_x,data_y)
    y_scores = cross_val_predict(classifier, data_x, data_y, cv=10)
    scores = cross_val_score(classifier, data_x, data_y, scoring='accuracy', cv=10)
    
    #graph ROC curve 
    fpr, tpr, thresholds= roc_curve(data_y,y_scores)
    save_plot_roc_curve(fpr, tpr, figname)
    
    return (precision_score(data_y, y_scores),recall_score(data_y, y_scores), roc_auc_score(data_y,y_scores), accuracy_score(data_y,y_scores))
```

```python

x_data = feature_selected_x_data()
counter = 1
for x in x_data:
    precision,recall,roc,acu = predict_with_data(x[0], Y.values.ravel(), createLogisticRegression(), x[1] + '_logreg')
    print(x[1] + '_logreg')
    print('Precision: ', precision)
    print('Recall: ', recall)
    print('ROC AUC Score: ', roc)
    print ('Accuracy: ', acu)
    # Recall, Precision, Accuracy
    print('---------')
    counter = counter + 1
    
    precision,recall,roc,acu = predict_with_data(x[0], Y.values.ravel(), createRandomForestClassifier(), x[1] + '_randfor')
    print(x[1] + '_randfor')
    print('Precision: ', precision)
    print('Recall: ', recall)
    print('ROC AUC Score: ', roc)
    print ('Accuracy: ', acu)
    # Recall, Precision, Accuracy
    print('---------')
    counter = counter + 1
    
    precision,recall,roc,acu = predict_with_data(x[0], Y.values.ravel(), createSVC(), x[1] + '_svc')
    print(x[1] + '_svc')
    print('Precision: ', precision)
    print('Recall: ', recall)
    print('ROC AUC Score: ', roc)
    print ('Accuracy: ', acu)
    # Recall, Precision, Accuracy
    print('---------')
    counter = counter + 1
```
